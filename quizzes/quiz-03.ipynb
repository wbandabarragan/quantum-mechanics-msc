## Problem 1 (6 points): Normalisation

A wave function for a particle in a box of length $L$ is given by:

\begin{equation*}
\psi(x) = C\,x(L-x)
\end{equation*}

(a) Find the normalisation constant, $C$.


(b) Why do wave functions need to be normalised?


(c) Using python, plot $\psi(x)$ versus $x$ (plug some fiducial values for the constants).


## Problem 2 (6 points): The time-independent Schrodinger equation
Consider the following 1D wave function:

\begin{equation*}
\psi(x) = A\,\left(\frac{x}{x_0}\right)^n\,e^{-\frac{x}{x_0}}
\end{equation*}

where $A$, $n$, and $x_0$ are constants.

(a) Write down the time-independent Schrodinger equation.


(b) Find the potential $V(x)$ and energy $E$, for which this wave function is a solution to the Schrodinger equation. Assume that as $x\to +\infty$, $V(x)\to 0$.

\end{equation}

(a) The expectation values associated with the position $x$: $\langle x \rangle$, $\langle x^2 \rangle$.


(b) The expectation values associated with the momentum $p$: $\langle p \rangle$, $\langle p^2 \rangle$.


(c) The dispersions $\sigma_x$ and $\sigma_p$, and their product $\sigma_x\,\sigma_p$.


(d) Use python to make a plot of $(\sigma_x\,\sigma_p)$ vs. $n$.


(e) Is the uncertainty principle satisfied? Which of the $\psi_n(x)$ states comes closest to the uncertainty limit?
